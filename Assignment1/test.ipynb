```python
class mycomplex:
    def __init__(self, real=0, imag=0):
        self.r = real
        self.i = imag
    def conjugate(self):
        return mycomplex(self.r, -self.i)
    def print(self):
        print(f"{self.r} + {self.i}i")
```

```python
y=mycomplex(3,4)
```

```python
y.print()
```

```python
z = y.conjugate()
```

```python
z.print()
```
