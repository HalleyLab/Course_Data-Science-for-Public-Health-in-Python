```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
kirb21_df = pd.read_csv("https://raw.githubusercontent.com/smart-stats/ds4bio_book/main/book/assetts/kirby21AllLevels.csv")
```

```python
kirby906a_ax_df = kirb21_df[kirb21_df['rawid'] == 'kirby906a_ax.img']
```

```python
telecephalon_L_volume = kirby906a_ax_df[(kirby906a_ax_df['roi'] == 'Telencephalon_L') & (kirby906a_ax_df['type'] == 1) & (kirby906a_ax_df['level'] == 1)]['volume'].values[0]
telecephalon_R_volume = kirby906a_ax_df[(kirby906a_ax_df['roi'] == 'Telencephalon_R') & (kirby906a_ax_df['type'] == 1) & (kirby906a_ax_df['level'] == 1)]['volume'].values[0]
print(telecephalon_L_volume)
print(telecephalon_R_volume)
```

```python
icv = kirby906a_ax_df['icv'].iloc[0]
tbv = kirby906a_ax_df['tbv'].iloc[0]
telecephalon_icv_fraction = (telecephalon_L_volume + telecephalon_R_volume) / icv
telecephalon_tbv_fraction = (telecephalon_L_volume + telecephalon_R_volume) / tbv
print("Fraction of ICV made up by Telencephalon is:", telecephalon_icv_fraction)
print("Fraction of TBV made up by Telencephalon is:", telecephalon_tbv_fraction)
```

```python
all_regions_df = kirby906a_ax_df.groupby(['type', 'level'])['volume'].sum().reset_index()
all_regions_df.rename(columns={'volume': 'total_volume'}, inplace=True)
print(all_regions_df) 
```

```python
type1_level2_df = kirby906a_ax_df[(kirby906a_ax_df['type'] == 1) & (kirby906a_ax_df['level'] == 2)]
type1_level2_df['combined_roi'] = type1_level2_df['roi'].str.replace('_L', '').str.replace('_R', '')
merged_t1l1_df = type1_level2_df.groupby('combined_roi', as_index=False)['volume'].sum()
plt.figure(figsize=(12, 6))
sns.barplot(data=merged_t1l1_df, x='combined_roi', y='volume')
plt.title("Volume Distribution (Type 1, Level 2)")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
type1_level3_df = kirby906a_ax_df[(kirby906a_ax_df['type'] == 1) & (kirby906a_ax_df['level'] == 3)]
type1_level3_df['combined_roi'] = type1_level3_df['roi'].str.replace('_L', '').str.replace('_R', '')
merged_t1l3_df = type1_level3_df.groupby('combined_roi', as_index=False)['volume'].sum().sample(frac=1, random_state=1).reset_index(drop=True)
plt.figure(figsize=(20, 20))
plt.pie(merged_t1l3_df['volume'], labels=merged_t1l3_df['combined_roi'], autopct='%1.1f%%', startangle=140, pctdistance=0.85)
plt.title("Volume Distribution (Type 1, Level 3)")
plt.rcParams['font.size'] = 12
centre_circle = plt.Circle((0,0), 0.50, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.tight_layout()
plt.show()
```

```python
healthcare_df = pd.read_csv("https://raw.githubusercontent.com/jhu-advdatasci/2018/master/data/KFF/healthcare-spending.csv", skiprows=2)
healthcare_df.dropna(inplace=True)
healthcare_df = healthcare_df[healthcare_df['Location'] != 'United States']
```

```python
healthcare_by_year_loc = healthcare_df.melt(id_vars='Location',var_name='Year',value_name='Total Health Spending')
healthcare_by_year_loc['Year'] = healthcare_by_year_loc['Year'].str.extract(r'(\d+)').astype(int)
```

```python
plt.figure(figsize=(20, 12))
sns.lineplot(data=healthcare_by_year_loc, x='Year', y='Total Health Spending', hue='Location')
plt.title("Healthcare Spending Over Time by State")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', ncol=2, fontsize='small')
plt.tight_layout()
plt.show()
```

```python
avg_spending_by_state = healthcare_by_year_loc.groupby('Location')['Total Health Spending'].mean().sort_values()
plt.figure(figsize=(16, 8))
plt.bar(avg_spending_by_state.index, avg_spending_by_state.values)
plt.title("Average Healthcare Spending by State")
plt.xlabel("State")
plt.ylabel("Average Spending")
plt.xticks(rotation=90, fontsize=8)
plt.tight_layout()
plt.show()
```
