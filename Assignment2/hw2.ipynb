```python
def sorting_hat():
    Q1 = "Are you selfless and brave?"
    Q2 = "Are you ambitious and cunning?"
    Q3 = "Are you value chivalry and loyalty?"
    A1 = input(Q1)
    A2 = input(Q2)
    if A1 == "Y" and A2 == "N":
        return "Gryffindor"
    elif A1 == "N" and A2 == "Y":
        return "Slytherin"
    else:
        A3 = input(Q3)
        if A3 == "Y":
            return "Gryffindor"
        else:
            return "Slytherin"
```

```python
class mydata:
    def __init__(self, data):
        self.data = data
    def mean(self):
        return sum(self.data) / len(self.data)
    def std(self):
        mean = self.mean()
        variance = sum((i - mean) ** 2 for i in self.data) / len(self.data)
        return variance ** 0.5
```

```python
def lag(vector, t):
    return np.concatenate((np.full(t, np.nan), vector[:-t]))
```

```python
def lagGenerator(t):
    return lambda vector: lag(vector, t)
```

```python
def filter_and_square(lst):
    return [i**2 for i in lst if i>0]
```

```python
def frequence_map(lst):
    freq_map = {}
    for item in lst:
        freq_map[item] = freq_map.get(item, 0) + 1
    return freq_map
```

```python
class extended_data(mydata):
    def min(self):
        min_val = self.data[0]
        for i in self.data:
            if i < min_val:
                min_val = i
        return min_val
    def max(self):
        max_val = self.data[0]
        for i in self.data:
            if i > max_val:
                max_val = i
        return max_val
```
